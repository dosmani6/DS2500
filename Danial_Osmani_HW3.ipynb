```python
class Node:
#defining one node in the whole class
    def __init__(self, name, category, dct=None):
        
        self.props = dct
        self.label = name
        self.category = category
        
    def retrieve_property(self, key):
        #
        if key in self.props:
            return self.props[key]  
        else:
            return None
        
    def retrieve_properties(self):
        #
        lst = []
        for k in self.props.keys():
            lst.append(self.get_property(k))
        return lst
            
    def set_property(self,value,key):
        #
        self.props[key] = value
    
    def __repr__(self):
        return str(self.label) +"|"+ str(self.category)
    
        
```
